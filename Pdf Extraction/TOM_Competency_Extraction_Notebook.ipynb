```python
import fitz  # PyMuPDF
import pandas as pd
```

```python
# PDF file name
pdf_file = "TOM competency_Disease competency_Nov2022_rev12Oct2023.pdf"

# List of Competencies and Proficiency Levels
competencies = [
    "Application of Epidemiological Principles", "Transboundary Diseases", "Disease prevention and control programmes",
    "Biosecurity", "Sampling", "Animal Identification, and Movement Control", "Emergency and Disaster Management",
    "Emergency Preparedness", "Emergency Response", "Veterinary products", "Animal Welfare", "Application of Risk Analysis",
    "Safety, Health and Wellbeing"
]
levels = ["Awareness", "Beginner", "Competent", "Proficient", "Expert"]

# Make an empty list to store extracted data
data = []

# Function to process the text and merge multi-line behaviors
def process_text(text):
    current_competency, current_level, current_behavior = None, None, ""
    lines = text.splitlines()

    for line in lines:
        line = line.strip()
        if line in competencies:
            if current_behavior:
                data.append([current_competency, current_level, current_behavior.strip()])
            current_competency, current_level, current_behavior = line, None, ""
        elif any(level in line for level in levels):
            if current_behavior:
                data.append([current_competency, current_level, current_behavior.strip()])
            current_level, current_behavior = next(level for level in levels if level in line), ""
        elif current_competency and current_level:
            if line.startswith("-"):
                if current_behavior:
                    data.append([current_competency, current_level, current_behavior.strip()])
                current_behavior = line
            else:
                current_behavior += f" {line}"
    if current_behavior:
        data.append([current_competency, current_level, current_behavior.strip()])

# Extract text from PDF using fitz (PyMuPDF)
def extract_text_from_pdf(pdf_file):
    with fitz.open(pdf_file) as pdf:
        text = "".join(page.get_text() for page in pdf)
    return text

# Extract text and process it
extracted_text = extract_text_from_pdf(pdf_file)
process_text(extracted_text)

# Create and save DataFrame from the extracted data
df = pd.DataFrame(data, columns=["Competency", "Level", "Behavior"])
df.to_csv("TOM_Competency_Framework_Extracted.csv", index=False, sep=";")

print("Extraction complete. Data saved to TOM_Competency_Framework_Extracted.csv")
```
