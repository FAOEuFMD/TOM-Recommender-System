```python
def create_df_pymysql(connection_details, query):
    try:
        # Establish the database connection
        conn = pymysql.connect(**connection_details)
    
        # Create the dataframe
        df = pd.read_sql(query, conn)
        
    except pymysql.Error as e:
        print(f"Error: {e}")
        
    finally:
        # Close the connection (VERY IMPORTANT!)
        if conn:
            conn.close()

    return df
```

```python
# combine the courses and the course category dataframe
merged_df = df_query_dict["courses"].join(df_query_dict["course_categories"].set_index('id'), on='category_id', how='left', lsuffix='_left', rsuffix='_right')
```

```python
courses_by_category_df = merged_df.groupby(['name_right']).count().reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# how many courses are there per category?

sns.barplot(data=courses_by_category_df, 
            x='id', 
            y='name_right',
            order=courses_by_category_df.sort_values('id',ascending = False)["name_right"]
           )
plt.title("Number of courses per category")
plt.xlabel("count")
plt.ylabel("course category")
plt.show()
```
