```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from rake_nltk import Rake
```

```python
courses = pd.read_excel("learning_objectives_main.xlsx")
courses
```

```python
courses['learning objectives']
```

```python
print(f"Total amount of NaNs : {courses['learning objectives'].isna().sum()}")
```

```python
courses = courses.dropna(subset=['learning objectives'])
print(f"Total amount NaNs, after cleaning : {courses['learning objectives'].isna().sum()}")
```

### First attempt for keywords

```python
def extract_keywords(text):
    if not isinstance(text, str):  # Make sure the input is a string
        return []  # Return an empty list for non-string values
    
    # Tokenize and filter
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    keywords = [word for word in tokens if word not in stop_words]  # Remove stop words
    
    return keywords  # Return the keywords

# Apply the function to extract keywords
courses['keywords'] = courses['learning objectives'].apply(extract_keywords)

# Convert keywords to a comma-separated string for better readability
courses['keywords_str'] = courses['keywords'].apply(lambda x: ', '.join(x))

# Let's have a look at the results
courses[['learning objectives', 'keywords_str']]
```

### Second attempt for keywords

A combination of the Rapid Automatic Keyword Extraction (RAKE) algorithm with the NLTK toolkit. It is known as rake-nltk. Find more information about it [HERE](https://pypi.org/project/rake-nltk/)

```python
rake_nltk_var = Rake()

def extract_keywords(text):
    if not isinstance(text, str):  # Ensure the input is a string
        return []  # Return an empty list for non-string values
   
    rake_nltk_var.extract_keywords_from_text(text)
    keyword_extracted = rake_nltk_var.get_ranked_phrases()
    
    return keyword_extracted  # Return as a list

# Apply the function to extract keywords
courses['keywords'] = courses['learning objectives'].apply(extract_keywords)

# Optional: Convert keywords to a comma-separated string for better readability
courses['keywords_2'] = courses['keywords'].apply(lambda x: ', '.join(x))

# Display results
courses[['learning objectives', 'keywords_2']]
```

### Comparison of both methods

```python
print(f"Original text : \n {courses['learning objectives'].iloc[876]}")
print("")
print(f"First method (removing stopwords) : \n{courses['keywords_str'].iloc[876]}")
print("")
print(f"Second method (using Rake) : \n{courses['keywords_2'].iloc[876]}")
```

**NOTES** : The first attempt seems to be giving better results
